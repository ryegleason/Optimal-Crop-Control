```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import linear_regression
```

```python
ziadi = pd.read_csv("ziadi_fig4.csv")
ziadi
```

```python
ziadi_filtered = ziadi[ziadi["NNI"] <= 1.0]
ziadi_filtered
```

```python
ziadi_filtered_shifted = ziadi_filtered - 1.0
ziadi_filtered_shifted
```

```python
ziadi_filtered_shifted["RY"]
```

```python
slope, intercept = linear_regression(ziadi_filtered_shifted["NNI"], ziadi_filtered_shifted["RY"], proportional=True)
slope
```

```python
intercept
```

```python
plt.scatter(ziadi_filtered["NNI"], ziadi_filtered["RY"])
x_series = np.linspace(min(ziadi_filtered["NNI"]), max(ziadi_filtered["NNI"]), 100)
plt.plot(x_series, (x_series - 1) * slope + 1)
```
